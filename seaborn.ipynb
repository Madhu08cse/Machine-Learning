```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.lineplot(x="total_bill",y="tip",data=tips)
plt.title("Line Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.barplot(x="total_bill",y="tip",data=tips)
plt.title("Bar Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.boxplot(x="total_bill",y="tip",data=tips)
plt.title("Box Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="tip",data=tips)
plt.title("Violin Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
correlation_matrix=iris.drop('species',axis=1).corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
sns.pairplot(iris)
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=iris,kind="hex")
plt.title("Joint Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
sns.countplot(x="sepal_length",data=iris)
plt.title("Count Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```
